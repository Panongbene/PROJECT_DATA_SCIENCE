```python
import numpy as np
import matplotlib.pyplot as plt
```

# Les cas triviaux

### Distribution Uniforme de Puissance $K_l$ fixe

```python
# Initialization values
B_l = 100000
K_l = 100
P_0 = 10
P_l = np.arange(P_0,200)
G_tilde = 24
Rho_tilde = 10
N_a = 1000
Sigma2 = 10
```

```python
def efficaciteEnergetique(B_l, K_l ,P_l ,G_tilde ,Rho_tilde ,N_a ,Sigma2):
    """This function compute the Energy efficience"""

    return (B_l*K_l/P_l)*np.log2(1+(P_l*N_a*Rho_tilde/K_l)/(P_l*(1+Rho_tilde)+Sigma2*G_tilde*Rho_tilde))
```

```python
epsilon = list()
for i in P_l:
    epsilon.append(efficaciteEnergetique(B_l, K_l, i, G_tilde, Rho_tilde, N_a, Sigma2))
```

```python
plt.figure(figsize=(20,15))
# Area plot
plt.xlabel('Puissance(P_l)')
plt.ylabel('Efficacité Energetique')
plt.title('Efficacité Energetique en fonction de la Puissance de la station de base')
plt.fill_between(P_l, epsilon)
#plt.savefig('fig1.png')
plt.show()
```

### Distribution Uniforme de Puissance $P_l$ fixe

```python
# Initialization values
B_l = 100000
K_l = np.arange(1,1000)
P_l = 150
G_tilde = 24
Rho_tilde = 10
N_a = 1000
Sigma2 = 10
```

```python
epsilon = list()
for i in K_l:
    epsilon.append(efficaciteEnergetique(B_l, i, P_l, G_tilde, Rho_tilde, N_a, Sigma2))
```

```python
plt.figure(figsize=(20,15))
# Area plot
plt.xlabel('Nombre d\'utilisateur(K_l)')
plt.ylabel('Efficacité Energetique')
plt.title('Efficacité Energetique en fonction du nombre d\'utilisateur servi par la station de base')
plt.fill_between(K_l, epsilon)
#plt.savefig('fig1.png')
plt.show()
```

### Distribution Uniforme de Puissance $P_l$ et $K_l$ variables

```python
# Initialization values
B_l = 100000
K_l = np.arange(1,500)
P_l = np.arange(P_0,200)
G_tilde = 24
Rho_tilde = 10
N_a = 1000
Sigma2 = 10
```

```python
X, Y = np.meshgrid(K_l, P_l)
 
# Résultat : tableau numpy 2D
Z = efficaciteEnergetique(B_l, X, Y, G_tilde, Rho_tilde, N_a, Sigma2)

# Affichage de la surface
plt.figure(figsize=(20,15))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')

# On légende la figure et on fixe une taille avec fontsize
ax.set_title("Efficacité énergétique en fonction de la puissance et du nombre d'utilisateur", fontsize = 13)
ax.set_xlabel('Nombre d\'utilisateur', fontsize = 11)
ax.set_ylabel('Puissance', fontsize = 11)
ax.set_zlabel('Efficacité énergétique', fontsize = 10)

plt.show()
```

```python
class ListeSimlpementChaineeFromScratch:
    """
        Cette classe implémente les listes simplement chaînée à partir de 0 dont les valeurs sont des entiers
    """
    
    def __init__(self, valeur_init_list):
        """
            Constructeur de notre classe
            @param valeur_init_list: valeur d'initialisation de notre liste qui est un entier
        """
        self.valeur = valeur_init_list
        self.next = None
        
    
    def __next__(self):
        """
            Cette méthode permet d'afficher le prochain élément de la liste
        """
        return self.next
    
    
    def __repr__(self):
        """
            Cette méthode est utilisée pour afficher notre liste
        """
        liste_sorti = []
        tamp = self
        while(tamp != None):
            liste_sorti.append(tamp.valeur)
            tamp = tamp.next
            
        return str(liste_sorti)

    
    def __len__(self):
        """
            Cette permet de retourner la longueur de la liste
        """
        tamp = self
        longueur_de_la_liste = 0
        if(tamp == None):
            return longueur_de_la_liste
        
        longueur_de_la_liste+=1
        while(tamp.next != None):
            tamp = tamp.next
            longueur_de_la_liste+=1
        
        return longueur_de_la_liste
    

    def append_new_value(self, new_valeur):
        """
            Cette méthode permet d'ajouter une nouvelle valeur à la fin de notre liste
            @param new_valeur: est la nouvelle valeur à ajouter
        """
        tamp = self
        while(tamp.next != None):
            tamp = tamp.next
        tamp.next = ListeSimlpementChaineeFromScratch(new_valeur)
        
        
        
    def remove_value_if_existe(self, valeur_to_remove):
        """
            Cette méthode permet la valeur en entrée dans la liste si cette valeur existe
            @param valeur_to_remove: est la valeur à supprimer
        """
        
        tamp = self
        if(tamp == None):
            return None
            
        while(tamp.next != None):
            
            if(tamp.next.valeur == valeur_to_remove):
                tamp.next = tamp.next.next
            else:
                tamp = tamp.next

        if(len(tamp) > 0):
            if(tamp.valeur == valeur_to_remove):
                self = None
                
    def __getitem__(self, index_value):
            """
                utiliser cette méthode pour acceder au valeur par index
                @param index_value: est un entier réprésentant l'index à acceder
            """
            
            if(index_value>=len(self)):
                return "--ERREUR : L'INDEX DEBORDE--"
            else:
                index_tampon = 0
                valeu_result = 0
                tamp = self
                while(tamp != None):
                    if(index_tampon == index_value):
                        valeu_result = tamp.valeur
                        break
                    index_tampon+=1
                    tamp = tamp.next
                return valeu_result
            
            
    def __setitem__(self, index_a_changer, nouvelle_valeur):
        """
            Cette méthode permet de changer la valeur d'un index
            @param index_a_changer: est l'index de la valeur à changer
            @param nouvelle_valeur: est la nouvelle valeur
        """
        if(index_a_changer>=len(self)):
            return "--ERREUR : L'INDEX DEBORDE--"
        else:
            index_tampon = 0
            tamp = self
            while(tamp != None):
                if(index_tampon == index_a_changer):
                    tamp.valeur = nouvelle_valeur
                    break
                index_tampon+=1
                tamp = tamp.next
                
    def __str__(self):
        """
            Cette méthode d'afficher la liste de manière plus jolie
        """
        
        result_affichage = "None--"
        tamp = self
        while(tamp != None):
            result_affichage += str(tamp.valeur)+" --> "
            tamp = tamp.next
        result_affichage+="None"
        return result_affichage.replace("None--", "")
    
    
    def __add__(self, list_to_add):
        """
            Cette permet d'implémenter l'operateur +
            @param list_to_add: est la liste à ajouter
        """
        tamp = self
        if(self == None):
            self = list_to_add
            return 
        tampon_result = ListeSimlpementChaineeFromScratch(tamp.valeur)
        
        tamp = self
        while(tamp.next is not None):
            tampon_result.append_new_value(tamp.next.valeur)
            tamp = tamp.next
            
        tamp = list_to_add
        while(tamp is not None):
            tampon_result.append_new_value(tamp.valeur)
            tamp = tamp.next
        return tampon_result
        
        
        
        
        
    def __iadd__(self, nouveau_liste_a_ajouter):
        """
            Cette permet l'utilisation de l'opperateur +=
            @param nouveau_liste_a_ajouter: est la liste à contacter 
        """
        """tamp = self
        if(tamp == None):
            self = nouveau_liste_a_ajouter 
            return """
        return self+nouveau_liste_a_ajouter        
```
